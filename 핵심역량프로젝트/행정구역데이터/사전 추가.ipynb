```python
import pandas as pd
import os

# 파일 경로 설정
os.chdir('C:/Users/USER/Desktop/핵심역량 프로젝트/데이터/행정구역 데이터/file/')

# 데이터 읽기
data = pd.read_csv('output_file.txt', header=None, sep='\t')

# 컬럼 이름 설정
data.columns = ['1', '2']

# 데이터 분할
data1 = data.iloc[:153962]
data2 = data.iloc[153962:]

# 중복 요소 제거
filtered_data2 = data2[~data2['1'].isin(data1['1'])]

# 중복 제거 완료 메시지
print("중복 요소 제거 완료!")

# 데이터 결합
result_data = pd.concat([data1, filtered_data2])

# 컬럼 값 합치기
result_data['combined'] = result_data['1'] + '\t' + result_data['2']

# 결과를 텍스트 파일로 저장 (큰 따옴표 및 역슬래시 없이)
with open('combined_output.txt', 'w', encoding='utf-8') as file:
    for line in result_data['combined']:
        file.write(line + '\n')
```
